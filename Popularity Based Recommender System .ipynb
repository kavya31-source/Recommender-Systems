```python
#Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading csv file
df_movies=pd.read_csv("movies.csv")
df_movies.head()
```

```python
df_movies.shape
```

```python
df_ratings=pd.read_csv("ratings.csv")
df_ratings.head()
```

```python
df_ratings.shape
```

```python
#Merging the 2 datasets on MovieID
df1=pd.merge(df_movies,df_ratings ,on=['movieId'],how='inner')
df1
```

```python
#Checking for null values
df1.isnull().sum()
```

```python
#Finding the number of users who rated for each movie and the mean rating
df2=df1.groupby('title').agg({'userId':'count','rating':'mean'})
df2.rename(columns={'userId':'ratingcount'},inplace=True)
```

```python
df2
```

```python
df2.describe()
```

```python
# We consider the movies with greater than 128 rating counts  
recommend_data=df2[df2['ratingcount']>139]
```

```python
# Finding the movies with the highest average ratings
top_5=recommend_data.sort_values(by='rating',ascending=False)
```

```python
#Top 5 movies to recommend with respect to their popularity
top_5.head()
```

```python
# User defined minimum ratings & count of movies 
movies=[]
ratings=int(input("Enter minimum rating:"))
count=int(input("Enter number of recommendation:"))
for i in top_5['rating']:
    if i>ratings:
        movies.append(top_5[top_5["rating"]==i].index[0])
        
pd.DataFrame(movies[0:count]).rename(columns={0:"TopMovies"})
```
