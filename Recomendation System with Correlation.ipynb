```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_book=pd.read_csv('book.csv',index_col=[0])
df_book.head()
```

```python
# Total ratings on each book.
df_book.groupby("bookTitle")[["bookRating"]].mean()
```

```python
# Total count of ratings on each book.
df_book.groupby("bookTitle")[["totalRatingCount"]].sum()
```

## Recomendation System Based on User Ratings :

```python
book_mat=df_book.pivot_table(index='userID',columns='bookTitle',values='bookRating')
book_mat.head()
```

##### Most rated Book :

```python
ratings=df_book.groupby('bookTitle').agg({'bookRating':'mean','totalRatingCount':'count'}).sort_values(by='totalRatingCount',ascending=False)
ratings
```

##### Choosing 2 movies : Wild Animus & The Lovely Bones: A Novel

```python
#User ratings for the above 2 movies
wa_user_rating=book_mat['Wild Animus']
wa_user_rating.head()
```

```python
lb_user_rating=book_mat['The Lovely Bones: A Novel']
lb_user_rating.head()
```

### Using Corrwith() to get correlation between 2 pandas series :

```python
similar_to_wa = book_mat.corrwith(wa_user_rating)
similar_to_lb = book_mat.corrwith(lb_user_rating)
```

```python
df_wa=pd.DataFrame(similar_to_wa,columns=['Correlation'])
df_wa.dropna(inplace=True)
df_wa.head()
```

```python
# Getting the correlation values based on the total number of ratings.
df_wa=df_wa.join(ratings['totalRatingCount'])
df_wa
```

```python
# Correlated book titles with respect to highest correlation
wa_rat=df_wa.sort_values('Correlation',ascending=False)
wa_rat
```

```python
# Considering the books only with totalRatingCount greater than 200.
top_5_1=wa_rat[wa_rat['totalRatingCount']>100].head()
top_5_1
```

##### The books recommended to the user after reading "Wild Animus" are :

```python
list(top_5_1.index)[1:]
```

##### Similarly for The Lovely Bones: A Novel :

```python
df_lb=pd.DataFrame(similar_to_lb,columns=['Correlation'])
df_lb.dropna(inplace=True)
df_lb.head()
```

```python
df_lb=df_lb.join(ratings['totalRatingCount'])
df_lb
```

```python
lb_rat=df_lb.sort_values('Correlation',ascending=False)
lb_rat
```

```python
top_5_2=lb_rat[lb_rat['totalRatingCount']>100].head()
top_5_2
```

##### The books recommended to the user after reading "The Lovely Bones: A Novel" are :

```python
list(top_5_2.index)[1:]
```
